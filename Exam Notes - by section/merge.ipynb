```python
# read in all .md files in the current directory and merge them into a single file
import os

# get all .md files in the current directory
md_files = [f for f in os.listdir('.') if f.endswith('.md')]

# sort alphanumerically (W101 before W903)
md_files.sort()

# read in all the files and add titles as headings
merged_sections = []
for f in md_files:
    # get the title (filename without .md extension)
    title = os.path.splitext(f)[0]
    # read the file content
    content = open(f, 'r').read()
    # add the title as a heading followed by the content
    merged_sections.append(f'# {title}\n\n{content}')

# merge the files
merged_content = '\n\n'.join(merged_sections)

# write the merged content to a new file
with open('merged.md', 'w') as f:
    f.write(merged_content)

```
